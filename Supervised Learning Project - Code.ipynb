##### Observations basis Classification Report:

```python
# Initially, we have used k=5 (5 nearest neighbors) to train the KNN Model. Now, we'll change it 3 and 10 respectively and determine their performance

knn3 = KNeighborsClassifier(n_neighbors = 3, weights = 'distance'). fit(x_train, y_train)
knn10 = KNeighborsClassifier(n_neighbors = 10, weights = 'distance'). fit(x_train, y_train)
```

```python
# Now, changing the weights from 'distance' to 'uniform', where the model assigns sames weight to all points irrespective of their distance
# Choosing the value for k=10 since this model has provided us with the best results previously

y_predict_uni = KNeighborsClassifier(n_neighbors=10, weights = 'uniform') .fit(x_test, y_test) .predict(x_test)
```

```python
# Creating Classification Report for each of the KNN Models with k=3, k=10 and k=5 (for comparison)

print(
    "Classification Report for KNN model with 10 nearest neighbors and 'uniform' distance: \n", classification_report(y_test, y_predict_uni)
)
```

#### Observations:

1. The Model's performance dropped signifcantly from 82% to 77% (-5 pp) due to a change in calculating/assigning weights for the class (based on 'distance' intially to uniformly weighing the nearest points to determine the class)
2. The worst affected seems to the 'Type_H' class whose precision, recall and f-1 score **dropped** by 23 pp, 64 pp and 58 pp respectively.

(pp -> percentage points)

#### Part 1.5 c - Identifying parameters contributed most to improve model performance

* Though only experimented with the number of nearest neighbors and the weights assigned to each of the points, we understand that the value of k and wights both had an impact on model's performance
* While k=3 and k=5 were facilitating creation of good models, the k=10 has the best performancing model.
* And changing the 'weights' parameter which initially assigned the weights basis Euclidian Distance, changing it to the 'uniform' created an adverse impact on our model.

#### Conclusion:

Basis the above details, the best performing model in my simulations has **_k=10_** value and **_weights = 'distance'_**

```python
# Finding the optimal 'k' value to the KNN Model

train_score = {}
test_score = {}
n_neighbors = np.arange(2, 30, 1)
for neighbor in n_neighbors:
    knn = KNeighborsClassifier(n_neighbors=neighbor, weights = 'distance')
    knn.fit(x_train, y_train)
    train_score[neighbor]=knn.score(x_train, y_train)
    test_score[neighbor]=knn.score(x_test, y_test)
```

```python
#Plotting the accuracy for each of 'k' values in KNN Model

plt.plot(n_neighbors, train_score.values(), label="Train Accuracy")
plt.plot(n_neighbors, test_score.values(), label="Test Accuracy")
plt.xlabel("Number Of Neighbors")
plt.ylabel("Accuracy")
plt.title("KNN: Varying number of Neighbors")
plt.legend()
plt.xlim(0, 33)
plt.ylim(0.0, 1.1)
plt.grid()
plt.show()
```

```python
#Printing the optimal 'k' value for the model
for k, value in test_score.items():
    if value==max(test_score.values()):
        print("k-value:", k)
```

####
#### Part 2.4 c- Printing evaluation metrics for final mode


```python
#Applying the optimal k value (k=27) to the base KNN Model;

knn27 = KNeighborsClassifier(n_neighbors = 27, weights = 'distance')

y_pred_k27 = knn27.fit(x_train,y_train) .predict(x_test)

# Creating a classification report for the KNN Model with k=7

k27_clfr = classification_report(y_test, y_pred_k27)
print(k27_clfr)
```
